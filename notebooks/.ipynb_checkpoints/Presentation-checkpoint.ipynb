```python
m = folium.Map([30.2672, -97.7431], zoom_start=11)
for index, row in coords.iterrows():
    folium.CircleMarker([row['lat'], row['lng']],
                        radius=15,
                        fill_color="#3db7e4",
                       ).add_to(m)
```
